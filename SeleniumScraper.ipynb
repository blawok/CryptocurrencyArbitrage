```python
browser = webdriver.Chrome(executable_path="C:\\Users\\Michał\\GIT\\CryptocurrencyArbitrage\\chromedriver.exe")
```

```python
browser.get("https://coinmarketcap.com/exchanges/")
```

```python
browser.find_elements_by_link_text("Top 100 By Reported Volume").click()
list_of_markets = []

```

```python
urls = []
infos = browser.find_elements_by_class_name('currency-name-container')
for url in infos:
    urls.append(url.get_attribute("href") + "#markets")

```

```python
exchange_lists = []
for url in urls[0:2]:
    browser.get(url)
    table_trs = browser.find_elements_by_xpath('//div[@id="markets"]/div/div[2]/div/table/tbody/tr')
    for tr in table_trs:
        single_exchange = []
        td = tr.find_elements_by_xpath(".//td")
        for i in [1,2,4]:
            single_exchange.append(td[i].get_attribute("data-sort"))
        exchange_lists.append(single_exchange)
```

```python
import pandas as pd

df = pd.DataFrame(exchange_lists)
```
